# **Datasets merge**

Merge the patients' information datasets to obtain a unique file from the two different data sources.

```python
random.seed(1234)
```

```python
patients_synthea = pd.read_csv(OUTPATH + "patients1.csv")
patients_emborg = pd.read_csv(OUTPATH + "patients2.csv")
```

```python
patients_unique = pd.merge(patients_synthea, patients_emborg, how="outer")
```

```python
patients_unique[1:5]
```

Generate random values for the gender, as the data coming fro synthea does not have this record. I chose to do so as the number of missing values is too high compared to the total number of occurences (1192/2092).

```python
gender = ["F","M"]
GENDER = random.choices(gender, k = patients_synthea.shape[0])

patients_unique["GENDER"].loc[:patients_synthea.shape[0] - 1 ] = GENDER
```

```python
patients_unique["GENDER"] = patients_unique["GENDER"].replace({"Male":"M", "Female":"F"})
```

```python
patients_unique.to_csv(OUTPATH + "patients.csv", index = False)
```
